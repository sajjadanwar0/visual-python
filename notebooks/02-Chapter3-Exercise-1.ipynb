```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
src_file=Path.cwd()/'data'/'raw'/'fuel.csv'
image_dir=Path.cwd()/'images'
```

```python
df=pd.read_csv(src_file)
df.info()
```

```python
%matplotlib inline
```

```python
plt.hist(df['comb08']);
```

```python
plt.hist(df['comb08'])
plt.xlabel('Combined Fuel Economy (MPG)')
plt.ylabel('Number of Cars')
plt.title('Combined Fuel Economy from 2000-2020')
plt.show()
```

Combine Boxplot with Histogram

```python
fig,ax=plt.subplots()
ax.boxplot(df['comb08'],labels=['All'])
ax.set(title='MGP distribution',ylabel='MPG for vehicles');

```

```python
fig,ax=plt.subplots()
ax.boxplot(df['comb08'],labels=['All'],showfliers=False)
ax.set(title='MGP distribution',ylabel='MPG for vehicles');
```

```python
fig,(ax1,ax2)=plt.subplots(2)
ax1.hist(df['comb08'],range=(10,50))
ax1.set(title='Vehicle MPG',xlabel='Combined MPG',ylabel='Number of Cars')
ax2.boxplot(df['comb08'],labels=['All'],vert=False,showfliers=False)
ax2.set(title='MGP distribution',ylabel='MPG for vehicles');

```

```python
fig,(ax1,ax2)=plt.subplots(2)
ax1.hist(df['comb08'],range=(10,50))
ax1.set(title='Vehicle MPG',xlabel='Combined MPG',ylabel='Number of Cars')
ax2.boxplot(df['comb08'],labels=['All'],vert=False,showfliers=False)
ax2.set(title='MGP distribution',ylabel='MPG for vehicles')

fig.suptitle('MGP Distribution',fontsize=14,fontweight='bold');
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(9,4))
ax1.hist(df['comb08'],range=(10,50))
ax1.set(title='Vehicle MPG',xlabel='Combined MPG',ylabel='Number of Cars')
ax2.boxplot(df['comb08'],labels=['All'],showfliers=False)
ax2.set(title='MGP distribution',ylabel='MPG for vehicles');
```
