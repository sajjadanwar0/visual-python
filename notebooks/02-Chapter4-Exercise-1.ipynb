```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
src_file = Path.cwd() / 'data' / 'raw' / 'fuel.csv'
```

```python
df = pd.read_csv(src_file)
```

```python
df.head()
```

```python
df['comb08'].plot(kind='hist');
```

```python
df['comb08'].plot.hist();
```

```python
df['comb08'].plot(kind='box');
```

```python
df['comb08'].plot(kind='density');
```

```python
df['comb08'].plot(kind='kde');
```

```python
avg_by_year = df.groupby(['year']).agg({'highway08': 'mean', 'city08': 'mean', 'comb08': 'mean'})
avg_by_year
```

```python
avg_by_year.plot(kind='box', title='Gas Mileage');
```

```python
avg_by_year.plot(kind='line', title='Gas Mileage');
```

```python
avg_by_year.plot(kind='line', figsize=(9, 5), xticks=np.arange(2000, 2022, 2)
                 , ylim=(0, 32), rot=45, title='Gas Mileage');
```

```python
avg_by_year.plot(kind='bar');
```

```python
avg_by_year.plot(kind='bar', rot=45, figsize=(10, 5));
```

```python
avg_by_year.plot(kind='barh');
```

```python
avg_by_year.plot(kind='area');
```

```python
fig, ax = plt.subplots()
avg_by_year.plot(kind='area', ax=ax);
```

```python
fig, ax1 = plt.subplots()
avg_by_year.plot(kind='area', ax=ax1, stacked=False);
ax1.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
ax1.set(ylabel='MPG', title='AvG MPG by year');
```
