```python
from pathlib import Path
import pandas as pd
import plotly.express as px
```

```python
src_file = Path.cwd() / 'data' / 'raw' / 'fuel_summary.csv'
df = pd.read_csv(src_file)
df.head()
```

```python
fig=px.strip(df,x="fuelCost08",y="class_summary",color="class_summary",range_x=[300,4000])
fig.update_layout(showlegend=False)
fig.update_yaxes(title="Vehicle Class")
fig.update_xaxes(title="Estimated Fuel Cost")
fig.show()
```

```python
fig=px.strip(df,x="fuelCost08",y="class_summary",color="class_summary",range_x=[300,4000])
fig.update_layout(showlegend=False)
fig.update_yaxes(title="Vehicle Class")
fig.update_xaxes(title="Estimated Fuel Cost")
fig.add_annotation(x=1150,y="Pickup",text="Outlier Value",showarrow=True,arrowhead=1)
fig.add_vrect(x0=3500,x1=4000,fillcolor="LightSalmon",
              opacity=0.5,layer="below",line_width=0,
              annotation_text="Inefficient Vehicles",
              annotation_position="bottom",
              )
fig.show()
fig.write_image("customization_example_plotly.svg")
```
