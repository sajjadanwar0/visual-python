```python
from pathlib import Path
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
src_file = Path.cwd() / 'data' / 'raw' / 'fuel_summary.csv'
df = pd.read_csv(src_file)
```

```python
df.head()
```

```python
sns.catplot(data=df,x='date_range',y='comb08',kind='box',showfliers=False)
```

```python
sns.catplot(data=df,x='date_range',y='comb08',kind='boxen')
```

```python
sns.catplot(data=df,x='date_range',y='comb08',kind='box',col='fuel_type_summary')
```

```python
sns.catplot(data=df,x='date_range',y='comb08',kind='box',col_wrap=2,col='fuel_type_summary')
```

```python
sns.catplot(data=df,x='year',y='highway08',aspect=2,kind='point')
```

```python
sns.catplot(data=df,x='year',y='highway08',hue='fuel_type_summary',join=False,aspect=2,kind='point')
```

```python
sns.catplot(data=df,x='year',y='highway08',aspect=2,row='date_range',hue='fuel_type_summary',sharex=False,kind='bar')
```

```python
sns.catplot(data=df,x='cylinders',y='comb08',col='drive',row='date_range',hue='transmission',kind='bar')
```
