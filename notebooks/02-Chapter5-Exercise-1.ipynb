### Searbon API: All searbon API has similar approache
like sns.catplot(data=df,x='comb08',y='cylinders')
we can also pass the row and columns like

col='drive'
row='date_range'

Also we can do the color things like hue
hue='transmission'

Type of plot: barplot---bar



```python
from pathlib import Path
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
src_file=Path.cwd()/'data'/'raw'/'fuel_summary.csv'
df=pd.read_csv(src_file)
```

```python
df.head()
```

```python
sns.set_style('darkgrid') #whitegrid,darkgrid
```

```python
sns.displot(df,x='comb08');
```

```python
sns.displot(data=df,x='comb08',kind='kde'); #kde or ecdf
```

```python
sns.displot(data=df,x='comb08',kind='ecdf'); #kde or ecdf
```

```python
sns.displot(data=df,x='comb08',kind='kde',rug=True);
```

```python
sns.histplot(data=df,x='comb08')
```

```python
sns.catplot(data=df,x='comb08')
```

```python
sns.stripplot(data=df,x='comb08')
```

```python
sns.catplot(data=df,x='comb08',kind='box')
```

```python
sns.catplot(data=df,y='comb08',kind='box')
```

```python
sns.catplot(data=df,x='comb08',kind='boxen')
```

```python
sns.catplot(data=df,y='comb08',kind='violin')
```

```python
sns.catplot(data=df,y='comb08',kind='bar')
```

```python
sns.catplot(data=df,x='cylinders',y='comb08',kind='bar')
```

```python
sns.catplot(data=df,x='cylinders',y='comb08',kind='point')
```

```python
sns.catplot(data=df,y='comb08',kind='point')
```

```python
sns.relplot(data=df,x='cylinders',y='comb08')
```

```python
sns.scatterplot(data=df,x='cylinders',y='comb08')
```
