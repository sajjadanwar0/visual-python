## Streamlit:
Launched in Oct 2019
##### Streamlit allows you to make interactive visualizations as simply as writing a python script.
##### There is relatively small API that enables the user to add the interactive elements to visualization tools of their choice(plotly,altair,
##### matplotlib,seaborn or others)

1) Build customs dashboards using pandas dataframe and multiple visualizations back-ends
2) Limited additional coding needed
3) Open source (Apache 2.0 license) is full-featured and maintained
4) Commercial offerings available from a venture-backed startup

#### Getting started
1) python -m pip install streamlit
2) Streamlit support its own native plot types
      - Line, bar charts
      - geographic charts

```python
from pathlib import Path
import pandas as pd
import plotly.express as px
import streamlit as st
```

```python
src_file = Path.cwd() / 'data' / 'raw' / 'fuel_summary.csv'
df = pd.read_csv(src_file)
df.head()
```

```python
fig=px.histogram(df,x="fuelCost08",color="class_summary",labels={"fuelCost":"Annual Fuel Cost"},
                 nbins=40,title="Fuel Cost Distribution"
                 )
st.title("Simple Example")
st.write(fig)

# Streamlit files are standard python files
# Use Standard Python tools (like pandas) to import and analyze data
# st.write() can display plots, images or D
```
